```python
import json
import numpy as np
from collections import defaultdict
import matplotlib.pyplot as plt
from BD import BD_PSNR, BD_RATE

def cal_metric(json_path):
    with open(json_path, 'r') as f:
        data = json.load(f)
    bpp = defaultdict()
    psnr = defaultdict()
    missim = defaultdict()
    for ds, results in data.items():
        bpp.setdefault(ds, defaultdict())
        psnr.setdefault(ds, defaultdict())
        missim.setdefault(ds, defaultdict())
    for ds, results in data.items():
        for seq_name, seq in results.items():
            for rate_num, res in seq.items():
                bpp[ds].setdefault(rate_num, []).append(np.mean(res['ave_all_frame_bpp']))
                psnr[ds].setdefault(rate_num, []).append(np.mean(res['ave_all_frame_psnr']))
                missim[ds].setdefault(rate_num, []).append(np.mean(res['ave_all_frame_msssim']))
    res = defaultdict()
    for ds, _ in data.items():
        res[ds] = defaultdict()
        for rate_num in bpp[ds].keys():
            res[ds].setdefault('bpp', []).append(np.mean(bpp[ds][rate_num]))
            res[ds].setdefault('psnr', []).append(np.mean(psnr[ds][rate_num]))
            res[ds].setdefault('msssim', []).append(np.mean(missim[ds][rate_num]))
#         for key, value in res.items():
#             res[key] = np.array(value)
    return res

base_path = './baseline/baseline.json'
json_path = './metric/dymask2.json'

baseline = cal_metric(base_path)
new_model = cal_metric(json_path)
for ds, res in new_model.items():
    plt.clf()
    plt.plot(baseline[ds]['bpp'], baseline[ds]['psnr'], label='baseline')
    plt.plot(new_model[ds]['bpp'], new_model[ds]['psnr'], label='new_model')
    plt.legend()
    plt.grid()
    plt.title(ds)
    plt.show()
    bd_psnr = BD_PSNR(baseline[ds]['bpp'], baseline[ds]['psnr'], new_model[ds]['bpp'], new_model[ds]['psnr'])
    bd_rate = BD_RATE(baseline[ds]['bpp'], baseline[ds]['psnr'], new_model[ds]['bpp'], new_model[ds]['psnr'])
    print('PSNR:', bd_psnr, 'dB', 'RATE:', bd_rate, '%')
    print('baseline bpp [', ','.join(list(map(str, np.around(baseline[ds]['bpp'], 4).tolist()))), ']')
    print('baseline psnr[', ','.join(list(map(str, np.around(baseline[ds]['psnr'], 3).tolist()))), ']')
    print('baseline ssim[', ','.join(list(map(str, np.around(baseline[ds]['msssim'], 3).tolist()))), ']')

    print('new bpp [', ','.join(list(map(str, np.around(new_model[ds]['bpp'], 4).tolist()))), ']')
    print('new psnr[', ','.join(list(map(str, np.around(new_model[ds]['psnr'], 3).tolist()))), ']')
    print('new ssim[', ','.join(list(map(str, np.around(new_model[ds]['msssim'], 3).tolist()))), ']')
```

```python
import matplotlib.pyplot as plt
import json
import numpy as np

sequeces = [
    "BasketballDrive/BasketballDrive",       
    "Cactus/Cactus",
    "Kimono1/Kimono1",
    "ParkScene/ParkScene",
    ]
def extract_seq_metric(json_path, seq_name='YachtRide/YachtRide', ds_name='JCT1080p'):
    with open(json_path, 'r') as f:
        data = json.load(f)
    bpp, psnr = [], []
    for rate_idx in ['000', '001', '002', '003']:
        bpp.append(data[ds_name][seq_name][rate_idx]['ave_all_frame_bpp'])
        psnr.append(data[ds_name][seq_name][rate_idx]['ave_all_frame_psnr'])
    return bpp, psnr

def plot_seq(seq):
    plt.figure(seq)
#     bpp_, psnr_ = extract_seq_metric(base_path, seq_name=seq)
    bpp, psnr = extract_seq_metric(json_path, seq_name=seq)
    fig = plt.figure()
#     plt.plot(bpp_, psnr_, label='baseline')
    plt.plot(bpp, psnr, label='new_model')
    plt.legend()
    plt.title(seq)
    plt.show()
    print('new bpp [', ','.join(list(map(str, np.around(bpp, 4).tolist()))), ']')
    print('new psnr [', ','.join(list(map(str, np.around(psnr, 4).tolist()))), ']')

for seq in sequeces:
    plot_seq(seq)
```

```python
import matplotlib.pyplot as plt
import json
import numpy as np

sequeces = [
    "Beauty/Beauty",       
    "Bosphorus/Bosphorus",
    "HoneyBee/HoneyBee",
    "ReadySteadyGo/ReadySteadyGo",
    "ShakeNDry/ShakeNDry",
    "Jockey/Jockey",
    "YachtRide/YachtRide",
    ]
def extract_metric(json_path, seq_name='YachtRide/YachtRide', rate_idx='003'):
    with open(json_path, 'r') as f:
        data = json.load(f)
    bpp = np.array(data['UVG'][seq_name][rate_idx]['frame_bpp'])
    psnr = np.array(data['UVG'][seq_name][rate_idx]['frame_psnr'])
    return bpp, psnr

def plot_seq(seq):
    bpp_, psnr_ = extract_metric(base_path, seq_name=seq)
    bpp, psnr = extract_metric(json_path, seq_name=seq)
    fig = plt.figure(figsize=(20, 10))
    markersize = 3
    plt.subplot(121)
    plt.plot(list(range(len(psnr_))), psnr_, marker='*', markersize=markersize, label="baseline")
    plt.plot(list(range(len(psnr))), psnr, marker='s', markersize=markersize, label="new_model")
    plt.legend()
    plt.title(seq)
#     plt.show()

#     fig = plt.figure()
    plt.subplot(122)
    plt.plot(list(range(len(bpp_))), bpp_, marker='*', markersize=markersize, label="baseline")
    plt.plot(list(range(len(bpp))), bpp, marker='s', markersize=markersize, label="new_model")
    plt.legend()
    plt.title(seq)
    plt.show()

for seq in sequeces:
    plot_seq(seq)
```
